```python
X=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
#X_train = scaler.transform(X_train)
#X_test = scaler.transform(X_test)
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='Adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
#smaller the batch size longer the training will take
#The batch_size argument in the model.fit() function in Keras specifies the number of samples that are processed before the model is updated. A larger batch size can improve the performance of the model, but it can also make the training process slower. A smaller batch size can make the training process faster, but it may not be as accurate.

#In the code you provided, the batch_size is set to 128. This means that the model will be updated after every 128 samples are processed. This is a good compromise between performance and speed.
```

```python
model.history.history
```

```python
df_loss = pd.DataFrame(model.history.history)
```

```python
df_loss
#loss is the training set loss and val_loss is the testing set loss
```

```python
df_loss.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```
